```python
import pandas as pd

# 加载训练集和测试集
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

# 查看数据基本信息
print('训练集形状:', train.shape)  # 输出：(891, 12)
print('测试集形状:', test.shape)    # 输出：(418, 11)

# 查看前几行数据
print(train.head())

# 查看数据类型和缺失值情况
print(train.info())
print(train.isnull().sum())  # 输出各字段缺失值数量
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Survived', data=train)
plt.title('幸存人数分布')
plt.show()
```

```python
# 例如：性别与幸存率
sns.barplot(x='Sex', y='Survived', data=train)
plt.title('性别对幸存率的影响')

# 船舱等级与幸存率
sns.barplot(x='Pclass', y='Survived', data=train)
plt.title('船舱等级对幸存率的影响')
```

```python
import matplotlib.pyplot as plt

# 先创建画布，设置更大的宽度
plt.figure(figsize=(10, 6))  # 宽 10，高 6（按需调整）
sns.barplot(x='Age_bins', y='Survived', data=train)
plt.xlabel('Age_bins（扩大画布）')
```

```python
# 示例：填充Age缺失值
train['Age'] = train['Age'].fillna(train['Age'].median())
test['Age'] = test['Age'].fillna(test['Age'].median())

# 填充Embarked缺失值
train['Embarked'] = train['Embarked'].fillna(train['Embarked'].mode()[0])
test['Fare'] = test['Fare'].fillna(test['Fare'].median())

# 处理Cabin：提取首字母
train['Cabin'] = train['Cabin'].str[0].fillna('U')  # U表示未知
test['Cabin'] = test['Cabin'].str[0].fillna('U')
```
